There are some discripancies in the values of "Education" and "Marriage", we see some extra values in those.

Conducting EDA on the basis of age:

```python
# Basic age statistics
print("Age Statistics:")
print(df['age'].describe())
```

```python
#making categories for different age groups
age_bins = pd.qcut(df['age'], q=5)
bin_counts = age_bins.value_counts().sort_index()

# Display the result
print(bin_counts)
```

```python
# Compute default rates for each quantile bin
default_rates = df.groupby(age_bins)['next_month_default'].mean() * 100

# Step 3: Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=default_rates.index.astype(str), y=default_rates.values, palette='Set3')

plt.ylabel('Default Rate (%)')
plt.xlabel('Age Group (Quantiles)')
plt.title('Default Rate by Age Quantile Group')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate default rates by age group
age_bins = [21, 27, 31, 37, 43, 79]
age_labels = ['21-27', '27-31', '31-37', '37-43', '43-79']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
default_by_age = df.groupby('age_group')['next_month_default'].mean() * 100

# Create an attractive pie chart
plt.figure(figsize=(12, 8))

# Custom color palette
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC']

# Explode the slices
explode = [0.05, 0.05, 0.05, 0.05, 0.05]

# Create the pie chart
plt.pie(default_by_age.values,
        labels=default_by_age.index,
        colors=colors,
        explode=explode,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90,
        textprops={'fontsize': 12, 'fontweight': 'bold'})

# Add title
plt.title('Default Rate Distribution Across Age Groups',
          pad=20,
          fontsize=16,
          fontweight='bold')

# Add legend
plt.legend(default_by_age.index,
          title="Age Groups",
          loc="center left",
          bbox_to_anchor=(1.1, 0.5))

plt.tight_layout()
plt.show()
```

```python
def custom_score(y_true, y_pred):
    f2 = fbeta_score(y_true, y_pred, beta=2)
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    
    final_score = 0.4 * f2 + 0.3 * acc + 0.3 * f1
    return final_score

# Create scorer object for hyperparameter tuning
custom_scorer = make_scorer(custom_score)

def hypertune_model(model, param_grid):
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])
    
    grid_search = GridSearchCV(
        pipeline,
        param_grid=param_grid,
        scoring=custom_scorer,
        cv=5,
        n_jobs=-1,
        verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    
    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_
```

```python
# Imports
import featuretools as ft
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score

# Define the custom weighted score function
def weighted_custom_score(y_true, y_pred):
    """Custom score: 0.6*F2 + 0.3*F1 + 0.1*Accuracy"""
    f2 = fbeta_score(y_true, y_pred, beta=2)
    f1 = f1_score(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    return 0.7 * f2 + 0.2 * f1 + 0.1 * acc

# Feature Engineering
es = ft.EntitySet(id='credit_data')
es.add_dataframe(dataframe_name='credit',
                 dataframe=df,
                 index='id',
                 make_index=True)

feature_matrix, feature_defs = ft.dfs(entityset=es,
                                    target_dataframe_name='credit',
                                    max_depth=2)

print("Generated Features:", len(feature_defs))
print("\nSample New Features:")
print(feature_matrix.columns.tolist()[:5])

# Data Preparation
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)

# Weighted Score Optimized Model
def build_weighted_score_model():
    model = Sequential([
        # Input Layer - optimized for F2/F1 balance
        Dense(1024, activation='relu', input_shape=(X_train.shape[1],), 
              kernel_regularizer=tf.keras.regularizers.l2(0.003)),
        BatchNormalization(),
        Dropout(0.45),
        
        # Hidden Layers - balanced for recall and precision
        Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.003)),
        BatchNormalization(),
        Dropout(0.35),
        
        Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.003)),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.003)),
        BatchNormalization(),
        Dropout(0.25),
        
        Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.003)),
        BatchNormalization(),
        Dropout(0.15),
        
        # Output Layer
        Dense(1, activation='sigmoid')
    ])
    
    # Optimized learning rate for weighted score
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0007)
    
    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Recall(), tf.keras.metrics.Precision()]
    )
    return model

# Weighted Score Focused Training Setup
class_weights = {0: 1, 1: 8}  # Strong focus on minority class for F2
y_train_smote = y_train_smote.reset_index(drop=True)
X_train_scaled = pd.DataFrame(X_train_scaled).reset_index(drop=True)

# Custom callbacks for weighted score optimization
callbacks = [
    EarlyStopping(monitor='val_recall', patience=25, restore_best_weights=True, mode='max'),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.25, patience=8, min_lr=1e-7),
    tf.keras.callbacks.ModelCheckpoint('best_weighted_model.keras', save_best_only=True, monitor='val_recall', mode='max')
]

# Model Training
model = build_weighted_score_model()
history = model.fit(
    X_train_scaled, y_train_smote,
    validation_split=0.2,
    epochs=200,  # Extended training for optimal convergence
    batch_size=24,  # Optimized batch size
    class_weight=class_weights,
    callbacks=callbacks
)

# Evaluation with multiple thresholds to find optimal
thresholds = [0.35, 0.4, 0.45, 0.5, 0.55, 0.85]
best_threshold = 0.5
best_score = 0

y_pred_proba = model.predict(X_test_scaled)

for threshold in thresholds:
    y_pred_thresh = (y_pred_proba > threshold).astype(int)
    score = weighted_custom_score(y_test, y_pred_thresh)
    print(f"Threshold {threshold}: Weighted Score = {score:.4f}")
    
    if score > best_score:
        best_score = score
        best_threshold = threshold

print(f"\n🎯 Best Threshold: {best_threshold} with Score: {best_score:.4f}")

# Final predictions with best threshold
y_pred_classes = (y_pred_proba > best_threshold).astype(int)

results = {
    "Model": "Weighted Score Optimized Neural Network",
    "Weighted Custom Score (0.7*F2 + 0.2*F1 + 0.1*Acc)": weighted_custom_score(y_test, y_pred_classes),
    "F2 Score": fbeta_score(y_test, y_pred_classes, beta=2),
    "F1 Score": f1_score(y_test, y_pred_classes),
    "Accuracy": accuracy_score(y_test, y_pred_classes),
    "Precision": precision_score(y_test, y_pred_classes),
    "Recall": recall_score(y_test, y_pred_classes)
}

# Display Results
print("\n🎯 WEIGHTED SCORE OPTIMIZED METRICS:")
for metric, value in results.items():
    if metric != "Model":
        print(f"{metric}: {value:.4f}")
    else:
        print(f"{metric}: {value}")

# Enhanced Visualization
plt.figure(figsize=(18, 6))

# Loss Plot
plt.subplot(1, 4, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Accuracy Plot
plt.subplot(1, 4, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# # Recall Plot
# plt.subplot(1, 4, 3)
# plt.plot(history.history['recall'], label='Training Recall')
# plt.plot(history.history['val_recall'], label='Validation Recall')
# plt.title('Model Recall (F2 Focus)')
# plt.xlabel('Epoch')
# plt.ylabel('Recall')
# plt.legend()

# # Precision Plot
# plt.subplot(1, 4, 4)
# plt.plot(history.history['precision'], label='Training Precision')
# plt.plot(history.history['val_precision'], label='Validation Precision')
# plt.title('Model Precision (F1 Balance)')
# plt.xlabel('Epoch')
# plt.ylabel('Precision')
# plt.legend()

plt.tight_layout()
plt.show()
```
